```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the data
mizrahi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [2.400, 2.680, 2.813, 2.103],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 25.0, 'Central Steppe': 24.8, 'Bactria-Margiana': 17.8,
            'Yellow River': 14.2, 'Indus Valley': 8.6, 'Copper Age Zagros': 4.2,
            'Central Siberian': 3.4, 'Arabian Peninsula': 1.4, 'Sub-Saharan African': 0.6
        },
        'Iron Age': {
            'Xiongnu': 44.0, 'Gandhara Grave': 19.4, 'Yaz Culture': 15.2,
            'Anatolian': 10.6, 'Saka': 9.8, 'Sub-Saharan African': 1.0
        },
        'Late Antiquity': {
            'Hunnic': 46.2, 'Khotanese Saka': 23.4, 'Iranian Plateau': 21.6,
            'Indian Subcontinent': 7.8, 'Sub-Saharan African': 1.0
        },
        'Middle Ages': {
            'Turkic': 48.0, 'Khorasan': 23.4, 'Mongolic': 22.8,
            'Iranian Plateau': 5.0, 'Sub-Saharan African': 0.8
        }
    }
}

ashkenazi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [3.789, 4.520, 5.192, 27.08],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 29.0, 'Western Steppe': 26.8, 'Bronze Age Caucasian': 21.4,
            'Yellow River': 15.2, 'Northwest African': 4.6, 'Canaanite': 2.4,
            'Central Siberian': 0.6
        },
        'Iron Age': {
            'Xiongnu': 47.2, 'Colchian': 31.2, 'Balto-Slavic': 13.8,
            'Sinitic': 4.8, 'Berber': 3.0
        },
        'Late Antiquity': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        },
        'Middle Ages': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        }
    }
}

def plot_genetic_fits_comparison(mizrahi_data, ashkenazi_data):
    """Plot genetic fits comparison across ages."""
    plt.figure(figsize=(12, 6))
    x = range(len(mizrahi_data['ages']))

    plt.plot(x, mizrahi_data['genetic_fits'], 'o-', label='Mizrahi')
    plt.plot(x, ashkenazi_data['genetic_fits'], 'o-', label='Ashkenazi')

    plt.xticks(x, mizrahi_data['ages'], rotation=45)
    plt.ylabel('Genetic Fit Score')
    plt.title('Comparison of Genetic Fits Across Time Periods')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def create_population_comparison_heatmap(mizrahi_data, ashkenazi_data, age):
    """Create a heatmap comparing population percentages for a specific age."""
    # Combine all unique populations
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]
    all_pops = list(set(list(miz_pops.keys()) + list(ash_pops.keys())))

    # Create comparison data
    comparison_data = []
    for pop in all_pops:
        miz_val = miz_pops.get(pop, 0)
        ash_val = ash_pops.get(pop, 0)
        comparison_data.append({
            'Population': pop,
            'Mizrahi': miz_val,
            'Ashkenazi': ash_val,
            'Difference': miz_val - ash_val
        })

    df = pd.DataFrame(comparison_data)
    df = df.sort_values('Difference', ascending=False)

    # Create heatmap
    plt.figure(figsize=(12, len(all_pops) * 0.5))
    sns.heatmap(df[['Mizrahi', 'Ashkenazi']].T,
                annot=True,
                fmt='.1f',
                cmap='RdYlBu',
                center=0,
                yticklabels=['Mizrahi', 'Ashkenazi'],
                xticklabels=df['Population'])
    plt.title(f'Population Percentages Comparison - {age}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    return df

def print_major_differences(comparison_df, age):
    """Print major differences in population percentages."""
    print(f"\nMajor differences in {age}:")
    for _, row in comparison_df.iterrows():
        if abs(row['Difference']) > 5:  # Only show differences > 5%
            print(f"{row['Population']}: " +
                  f"Mizrahi ({row['Mizrahi']:.1f}%) vs " +
                  f"Ashkenazi ({row['Ashkenazi']:.1f}%) = " +
                  f"{row['Difference']:.1f}% difference")

# Execute the analysis
def main():
    # Plot genetic fits comparison
    plot_genetic_fits_comparison(mizrahi_data, ashkenazi_data)

    # Analyze each time period
    for age in mizrahi_data['ages']:
        print(f"\n{'='*50}")
        print(f"Analysis for {age}")
        print('='*50)

        comparison_df = create_population_comparison_heatmap(mizrahi_data, ashkenazi_data, age)
        print_major_differences(comparison_df, age)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Circle, ConnectionPatch
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# [Previous data definitions remain the same]
mizrahi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [2.400, 2.680, 2.813, 2.103],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 25.0, 'Central Steppe': 24.8, 'Bactria-Margiana': 17.8,
            'Yellow River': 14.2, 'Indus Valley': 8.6, 'Copper Age Zagros': 4.2,
            'Central Siberian': 3.4, 'Arabian Peninsula': 1.4, 'Sub-Saharan African': 0.6
        },
        'Iron Age': {
            'Xiongnu': 44.0, 'Gandhara Grave': 19.4, 'Yaz Culture': 15.2,
            'Anatolian': 10.6, 'Saka': 9.8, 'Sub-Saharan African': 1.0
        },
        'Late Antiquity': {
            'Hunnic': 46.2, 'Khotanese Saka': 23.4, 'Iranian Plateau': 21.6,
            'Indian Subcontinent': 7.8, 'Sub-Saharan African': 1.0
        },
        'Middle Ages': {
            'Turkic': 48.0, 'Khorasan': 23.4, 'Mongolic': 22.8,
            'Iranian Plateau': 5.0, 'Sub-Saharan African': 0.8
        }
    }
}

ashkenazi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [3.789, 4.520, 5.192, 27.08],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 29.0, 'Western Steppe': 26.8, 'Bronze Age Caucasian': 21.4,
            'Yellow River': 15.2, 'Northwest African': 4.6, 'Canaanite': 2.4,
            'Central Siberian': 0.6
        },
        'Iron Age': {
            'Xiongnu': 47.2, 'Colchian': 31.2, 'Balto-Slavic': 13.8,
            'Sinitic': 4.8, 'Berber': 3.0
        },
        'Late Antiquity': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        },
        'Middle Ages': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        }
    }
}

def create_radar_plot(mizrahi_data, ashkenazi_data, age):
    """Create a radar plot comparing population percentages."""
    # Get all populations for the specific age
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]
    all_pops = list(set(list(miz_pops.keys()) + list(ash_pops.keys())))

    # Prepare data
    miz_values = [miz_pops.get(pop, 0) for pop in all_pops]
    ash_values = [ash_pops.get(pop, 0) for pop in all_pops]

    # Create the radar plot
    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=miz_values,
        theta=all_pops,
        fill='toself',
        name='Mizrahi'
    ))

    fig.add_trace(go.Scatterpolar(
        r=ash_values,
        theta=all_pops,
        fill='toself',
        name='Ashkenazi'
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, max(max(miz_values), max(ash_values))]
            )),
        showlegend=True,
        title=f"Population Comparison - {age}"
    )

    return fig

def create_sankey_diagram(mizrahi_data, ashkenazi_data, age):
    """Create a Sankey diagram showing population flows."""
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]

    # Prepare nodes
    nodes = ['Mizrahi', 'Ashkenazi'] + list(set(list(miz_pops.keys()) + list(ash_pops.keys())))
    node_indices = {node: idx for idx, node in enumerate(nodes)}

    # Prepare links
    source = []
    target = []
    value = []

    # Add Mizrahi connections
    for pop, percent in miz_pops.items():
        source.append(node_indices['Mizrahi'])
        target.append(node_indices[pop])
        value.append(percent)

    # Add Ashkenazi connections
    for pop, percent in ash_pops.items():
        source.append(node_indices['Ashkenazi'])
        target.append(node_indices[pop])
        value.append(percent)

    # Create the Sankey diagram
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=nodes,
            color="blue"
        ),
        link=dict(
            source=source,
            target=target,
            value=value
        )
    )])

    fig.update_layout(
        title_text=f"Genetic Flow Diagram - {age}",
        font_size=10
    )

    return fig

def create_bubble_plot(mizrahi_data, ashkenazi_data):
    """Create an interactive bubble plot showing genetic fits and population sizes."""
    data = []

    for age in mizrahi_data['ages']:
        # Calculate total population percentage for size reference
        miz_total = sum(mizrahi_data['populations'][age].values())
        ash_total = sum(ashkenazi_data['populations'][age].values())

        # Add Mizrahi data point
        data.append({
            'Age': age,
            'Group': 'Mizrahi',
            'Genetic Fit': mizrahi_data['genetic_fits'][mizrahi_data['ages'].index(age)],
            'Total Population': miz_total,
            'Dominant Population': max(mizrahi_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

        # Add Ashkenazi data point
        data.append({
            'Age': age,
            'Group': 'Ashkenazi',
            'Genetic Fit': ashkenazi_data['genetic_fits'][ashkenazi_data['ages'].index(age)],
            'Total Population': ash_total,
            'Dominant Population': max(ashkenazi_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

    df = pd.DataFrame(data)

    fig = px.scatter(df,
                    x='Age',
                    y='Genetic Fit',
                    size='Total Population',
                    color='Group',
                    hover_data=['Dominant Population'],
                    title='Genetic Fits and Population Composition Over Time')

    return fig

def plot_all_visualizations():
    """Create and display all visualizations."""
    # Create bubble plot
    bubble_fig = create_bubble_plot(mizrahi_data, ashkenazi_data)
    bubble_fig.show()

    # Create visualizations for each time period
    for age in mizrahi_data['ages']:
        # Radar plot
        radar_fig = create_radar_plot(mizrahi_data, ashkenazi_data, age)
        radar_fig.show()

        # Sankey diagram
        sankey_fig = create_sankey_diagram(mizrahi_data, ashkenazi_data, age)
        sankey_fig.show()

if __name__ == "__main__":
    # Make sure to install required packages
    # !pip install plotly
    plot_all_visualizations()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Circle, ConnectionPatch
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# [Previous Mizrahi and Ashkenazi data definitions remain the same]
mizrahi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [2.400, 2.680, 2.813, 2.103],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 25.0, 'Central Steppe': 24.8, 'Bactria-Margiana': 17.8,
            'Yellow River': 14.2, 'Indus Valley': 8.6, 'Copper Age Zagros': 4.2,
            'Central Siberian': 3.4, 'Arabian Peninsula': 1.4, 'Sub-Saharan African': 0.6
        },
        'Iron Age': {
            'Xiongnu': 44.0, 'Gandhara Grave': 19.4, 'Yaz Culture': 15.2,
            'Anatolian': 10.6, 'Saka': 9.8, 'Sub-Saharan African': 1.0
        },
        'Late Antiquity': {
            'Hunnic': 46.2, 'Khotanese Saka': 23.4, 'Iranian Plateau': 21.6,
            'Indian Subcontinent': 7.8, 'Sub-Saharan African': 1.0
        },
        'Middle Ages': {
            'Turkic': 48.0, 'Khorasan': 23.4, 'Mongolic': 22.8,
            'Iranian Plateau': 5.0, 'Sub-Saharan African': 0.8
        }
    }
}

ashkenazi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [3.789, 4.520, 5.192, 27.08],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 29.0, 'Western Steppe': 26.8, 'Bronze Age Caucasian': 21.4,
            'Yellow River': 15.2, 'Northwest African': 4.6, 'Canaanite': 2.4,
            'Central Siberian': 0.6
        },
        'Iron Age': {
            'Xiongnu': 47.2, 'Colchian': 31.2, 'Balto-Slavic': 13.8,
            'Sinitic': 4.8, 'Berber': 3.0
        },
        'Late Antiquity': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        },
        'Middle Ages': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        }
    }
}

# Add Sephardic data
sephardic_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [3.789, 5.393, 5.192, 26.662],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 29.0, 'Western Steppe': 26.8, 'Bronze Age Caucasian': 21.4,
            'Yellow River': 15.2, 'Northwest African': 4.6, 'Canaanite': 2.4,
            'Central Siberian': 0.6
        },
        'Iron Age': {
            'Xiongnu': 53.4, 'Colchian': 34.6, 'Italic and Etruscan': 9.0,
            'Berber': 2.6, 'Sinitic': 0.4
        },
        'Late Antiquity': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        },
        'Middle Ages': {
            'Slavic': 66.2, 'North African': 26.4, 'European Jew': 7.4
        }
    }
}

def create_radar_plot(mizrahi_data, ashkenazi_data, sephardic_data, age):
    """Create a radar plot comparing population percentages."""
    # Get all populations for the specific age
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]
    sep_pops = sephardic_data['populations'][age]
    all_pops = list(set(list(miz_pops.keys()) + list(ash_pops.keys()) + list(sep_pops.keys())))

    # Prepare data
    miz_values = [miz_pops.get(pop, 0) for pop in all_pops]
    ash_values = [ash_pops.get(pop, 0) for pop in all_pops]
    sep_values = [sep_pops.get(pop, 0) for pop in all_pops]

    # Create the radar plot
    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=miz_values,
        theta=all_pops,
        fill='toself',
        name='Mizrahi'
    ))

    fig.add_trace(go.Scatterpolar(
        r=ash_values,
        theta=all_pops,
        fill='toself',
        name='Ashkenazi'
    ))

    fig.add_trace(go.Scatterpolar(
        r=sep_values,
        theta=all_pops,
        fill='toself',
        name='Sephardic'
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, max(max(miz_values), max(ash_values), max(sep_values))]
            )),
        showlegend=True,
        title=f"Population Comparison - {age}"
    )

    return fig

def create_sankey_diagram(mizrahi_data, ashkenazi_data, sephardic_data, age):
    """Create a Sankey diagram showing population flows."""
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]
    sep_pops = sephardic_data['populations'][age]

    # Prepare nodes
    nodes = ['Mizrahi', 'Ashkenazi', 'Sephardic'] + list(set(list(miz_pops.keys()) + list(ash_pops.keys()) + list(sep_pops.keys())))
    node_indices = {node: idx for idx, node in enumerate(nodes)}

    # Prepare links
    source = []
    target = []
    value = []

    # Add Mizrahi connections
    for pop, percent in miz_pops.items():
        source.append(node_indices['Mizrahi'])
        target.append(node_indices[pop])
        value.append(percent)

    # Add Ashkenazi connections
    for pop, percent in ash_pops.items():
        source.append(node_indices['Ashkenazi'])
        target.append(node_indices[pop])
        value.append(percent)

    # Add Sephardic connections
    for pop, percent in sep_pops.items():
        source.append(node_indices['Sephardic'])
        target.append(node_indices[pop])
        value.append(percent)

    # Create the Sankey diagram
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=nodes,
            color=["#1f77b4", "#ff7f0e", "#2ca02c"] + ["#7f7f7f"] * len(nodes[3:])  # Different colors for Jewish groups
        ),
        link=dict(
            source=source,
            target=target,
            value=value
        )
    )])

    fig.update_layout(
        title_text=f"Genetic Flow Diagram - {age}",
        font_size=10
    )

    return fig

def create_bubble_plot(mizrahi_data, ashkenazi_data, sephardic_data):
    """Create an interactive bubble plot showing genetic fits and population sizes."""
    data = []

    for age in mizrahi_data['ages']:
        # Calculate total population percentage for size reference
        miz_total = sum(mizrahi_data['populations'][age].values())
        ash_total = sum(ashkenazi_data['populations'][age].values())
        sep_total = sum(sephardic_data['populations'][age].values())

        # Add Mizrahi data point
        data.append({
            'Age': age,
            'Group': 'Mizrahi',
            'Genetic Fit': mizrahi_data['genetic_fits'][mizrahi_data['ages'].index(age)],
            'Total Population': miz_total,
            'Dominant Population': max(mizrahi_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

        # Add Ashkenazi data point
        data.append({
            'Age': age,
            'Group': 'Ashkenazi',
            'Genetic Fit': ashkenazi_data['genetic_fits'][ashkenazi_data['ages'].index(age)],
            'Total Population': ash_total,
            'Dominant Population': max(ashkenazi_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

        # Add Sephardic data point
        data.append({
            'Age': age,
            'Group': 'Sephardic',
            'Genetic Fit': sephardic_data['genetic_fits'][mizrahi_data['ages'].index(age)],
            'Total Population': sep_total,
            'Dominant Population': max(sephardic_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

    df = pd.DataFrame(data)

    fig = px.scatter(df,
                    x='Age',
                    y='Genetic Fit',
                    size='Total Population',
                    color='Group',
                    hover_data=['Dominant Population'],
                    title='Genetic Fits and Population Composition Over Time')

    # Update y-axis to use log scale due to large range in genetic fits
    fig.update_layout(yaxis_type="log")

    return fig

def create_stacked_bar_comparison(mizrahi_data, ashkenazi_data, sephardic_data, age):
    """Create a stacked bar chart comparing major populations."""
    fig = go.Figure()

    groups = ['Mizrahi', 'Ashkenazi', 'Sephardic']
    data_dict = {
        'Mizrahi': mizrahi_data['populations'][age],
        'Ashkenazi': ashkenazi_data['populations'][age],
        'Sephardic': sephardic_data['populations'][age]
    }

    # Sort populations by maximum percentage across all groups
    all_pops = set()
    for d in data_dict.values():
        all_pops.update(d.keys())

    max_values = {pop: max(data_dict[g].get(pop, 0) for g in groups) for pop in all_pops}
    sorted_pops = sorted(all_pops, key=lambda x: max_values[x], reverse=True)

    # Create stacked bars
    for pop in sorted_pops:
        values = [data_dict[g].get(pop, 0) for g in groups]
        fig.add_trace(go.Bar(
            name=pop,
            x=groups,
            y=values,
            text=values,
            textposition='auto',
        ))

    fig.update_layout(
        barmode='stack',
        title=f"Population Distribution Comparison - {age}",
        yaxis_title="Percentage",
        showlegend=True
    )

    return fig

def plot_all_visualizations():
    """Create and display all visualizations."""
    # Create bubble plot
    bubble_fig = create_bubble_plot(mizrahi_data, ashkenazi_data, sephardic_data)
    bubble_fig.show()

    # Create visualizations for each time period
    for age in mizrahi_data['ages']:
        # Radar plot
        radar_fig = create_radar_plot(mizrahi_data, ashkenazi_data, sephardic_data, age)
        radar_fig.show()

        # Sankey diagram
        sankey_fig = create_sankey_diagram(mizrahi_data, ashkenazi_data, sephardic_data, age)
        sankey_fig.show()

        # Stacked bar chart
        stacked_fig = create_stacked_bar_comparison(mizrahi_data, ashkenazi_data, sephardic_data, age)
        stacked_fig.show()

if __name__ == "__main__":
    # Make sure to install required packages
    # !pip install plotly
    plot_all_visualizations()
```
