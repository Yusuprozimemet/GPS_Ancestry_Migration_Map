```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the data
mizrahi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [2.400, 2.680, 2.813, 2.103],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 25.0, 'Central Steppe': 24.8, 'Bactria-Margiana': 17.8,
            'Yellow River': 14.2, 'Indus Valley': 8.6, 'Copper Age Zagros': 4.2,
            'Central Siberian': 3.4, 'Arabian Peninsula': 1.4, 'Sub-Saharan African': 0.6
        },
        'Iron Age': {
            'Xiongnu': 44.0, 'Gandhara Grave': 19.4, 'Yaz Culture': 15.2,
            'Anatolian': 10.6, 'Saka': 9.8, 'Sub-Saharan African': 1.0
        },
        'Late Antiquity': {
            'Hunnic': 46.2, 'Khotanese Saka': 23.4, 'Iranian Plateau': 21.6,
            'Indian Subcontinent': 7.8, 'Sub-Saharan African': 1.0
        },
        'Middle Ages': {
            'Turkic': 48.0, 'Khorasan': 23.4, 'Mongolic': 22.8,
            'Iranian Plateau': 5.0, 'Sub-Saharan African': 0.8
        }
    }
}

ashkenazi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [3.789, 4.520, 5.192, 27.08],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 29.0, 'Western Steppe': 26.8, 'Bronze Age Caucasian': 21.4,
            'Yellow River': 15.2, 'Northwest African': 4.6, 'Canaanite': 2.4,
            'Central Siberian': 0.6
        },
        'Iron Age': {
            'Xiongnu': 47.2, 'Colchian': 31.2, 'Balto-Slavic': 13.8,
            'Sinitic': 4.8, 'Berber': 3.0
        },
        'Late Antiquity': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        },
        'Middle Ages': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        }
    }
}

def plot_genetic_fits_comparison(mizrahi_data, ashkenazi_data):
    """Plot genetic fits comparison across ages."""
    plt.figure(figsize=(12, 6))
    x = range(len(mizrahi_data['ages']))

    plt.plot(x, mizrahi_data['genetic_fits'], 'o-', label='Mizrahi')
    plt.plot(x, ashkenazi_data['genetic_fits'], 'o-', label='Ashkenazi')

    plt.xticks(x, mizrahi_data['ages'], rotation=45)
    plt.ylabel('Genetic Fit Score')
    plt.title('Comparison of Genetic Fits Across Time Periods')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def create_population_comparison_heatmap(mizrahi_data, ashkenazi_data, age):
    """Create a heatmap comparing population percentages for a specific age."""
    # Combine all unique populations
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]
    all_pops = list(set(list(miz_pops.keys()) + list(ash_pops.keys())))

    # Create comparison data
    comparison_data = []
    for pop in all_pops:
        miz_val = miz_pops.get(pop, 0)
        ash_val = ash_pops.get(pop, 0)
        comparison_data.append({
            'Population': pop,
            'Mizrahi': miz_val,
            'Ashkenazi': ash_val,
            'Difference': miz_val - ash_val
        })

    df = pd.DataFrame(comparison_data)
    df = df.sort_values('Difference', ascending=False)

    # Create heatmap
    plt.figure(figsize=(12, len(all_pops) * 0.5))
    sns.heatmap(df[['Mizrahi', 'Ashkenazi']].T,
                annot=True,
                fmt='.1f',
                cmap='RdYlBu',
                center=0,
                yticklabels=['Mizrahi', 'Ashkenazi'],
                xticklabels=df['Population'])
    plt.title(f'Population Percentages Comparison - {age}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    return df

def print_major_differences(comparison_df, age):
    """Print major differences in population percentages."""
    print(f"\nMajor differences in {age}:")
    for _, row in comparison_df.iterrows():
        if abs(row['Difference']) > 5:  # Only show differences > 5%
            print(f"{row['Population']}: " +
                  f"Mizrahi ({row['Mizrahi']:.1f}%) vs " +
                  f"Ashkenazi ({row['Ashkenazi']:.1f}%) = " +
                  f"{row['Difference']:.1f}% difference")

# Execute the analysis
def main():
    # Plot genetic fits comparison
    plot_genetic_fits_comparison(mizrahi_data, ashkenazi_data)

    # Analyze each time period
    for age in mizrahi_data['ages']:
        print(f"\n{'='*50}")
        print(f"Analysis for {age}")
        print('='*50)

        comparison_df = create_population_comparison_heatmap(mizrahi_data, ashkenazi_data, age)
        print_major_differences(comparison_df, age)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Circle, ConnectionPatch
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# [Previous data definitions remain the same]
mizrahi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [2.400, 2.680, 2.813, 2.103],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 25.0, 'Central Steppe': 24.8, 'Bactria-Margiana': 17.8,
            'Yellow River': 14.2, 'Indus Valley': 8.6, 'Copper Age Zagros': 4.2,
            'Central Siberian': 3.4, 'Arabian Peninsula': 1.4, 'Sub-Saharan African': 0.6
        },
        'Iron Age': {
            'Xiongnu': 44.0, 'Gandhara Grave': 19.4, 'Yaz Culture': 15.2,
            'Anatolian': 10.6, 'Saka': 9.8, 'Sub-Saharan African': 1.0
        },
        'Late Antiquity': {
            'Hunnic': 46.2, 'Khotanese Saka': 23.4, 'Iranian Plateau': 21.6,
            'Indian Subcontinent': 7.8, 'Sub-Saharan African': 1.0
        },
        'Middle Ages': {
            'Turkic': 48.0, 'Khorasan': 23.4, 'Mongolic': 22.8,
            'Iranian Plateau': 5.0, 'Sub-Saharan African': 0.8
        }
    }
}

ashkenazi_data = {
    'ages': ['Bronze Age', 'Iron Age', 'Late Antiquity', 'Middle Ages'],
    'genetic_fits': [3.789, 4.520, 5.192, 27.08],
    'populations': {
        'Bronze Age': {
            'Eastern Steppe': 29.0, 'Western Steppe': 26.8, 'Bronze Age Caucasian': 21.4,
            'Yellow River': 15.2, 'Northwest African': 4.6, 'Canaanite': 2.4,
            'Central Siberian': 0.6
        },
        'Iron Age': {
            'Xiongnu': 47.2, 'Colchian': 31.2, 'Balto-Slavic': 13.8,
            'Sinitic': 4.8, 'Berber': 3.0
        },
        'Late Antiquity': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        },
        'Middle Ages': {
            'Hunnic': 54.2, 'Roman Levant': 26.2, 'Slavic': 15.2, 'Sinitic': 4.4
        }
    }
}

def create_radar_plot(mizrahi_data, ashkenazi_data, age):
    """Create a radar plot comparing population percentages."""
    # Get all populations for the specific age
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]
    all_pops = list(set(list(miz_pops.keys()) + list(ash_pops.keys())))

    # Prepare data
    miz_values = [miz_pops.get(pop, 0) for pop in all_pops]
    ash_values = [ash_pops.get(pop, 0) for pop in all_pops]

    # Create the radar plot
    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=miz_values,
        theta=all_pops,
        fill='toself',
        name='Mizrahi'
    ))

    fig.add_trace(go.Scatterpolar(
        r=ash_values,
        theta=all_pops,
        fill='toself',
        name='Ashkenazi'
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, max(max(miz_values), max(ash_values))]
            )),
        showlegend=True,
        title=f"Population Comparison - {age}"
    )

    return fig

def create_sankey_diagram(mizrahi_data, ashkenazi_data, age):
    """Create a Sankey diagram showing population flows."""
    miz_pops = mizrahi_data['populations'][age]
    ash_pops = ashkenazi_data['populations'][age]

    # Prepare nodes
    nodes = ['Mizrahi', 'Ashkenazi'] + list(set(list(miz_pops.keys()) + list(ash_pops.keys())))
    node_indices = {node: idx for idx, node in enumerate(nodes)}

    # Prepare links
    source = []
    target = []
    value = []

    # Add Mizrahi connections
    for pop, percent in miz_pops.items():
        source.append(node_indices['Mizrahi'])
        target.append(node_indices[pop])
        value.append(percent)

    # Add Ashkenazi connections
    for pop, percent in ash_pops.items():
        source.append(node_indices['Ashkenazi'])
        target.append(node_indices[pop])
        value.append(percent)

    # Create the Sankey diagram
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=nodes,
            color="blue"
        ),
        link=dict(
            source=source,
            target=target,
            value=value
        )
    )])

    fig.update_layout(
        title_text=f"Genetic Flow Diagram - {age}",
        font_size=10
    )

    return fig

def create_bubble_plot(mizrahi_data, ashkenazi_data):
    """Create an interactive bubble plot showing genetic fits and population sizes."""
    data = []

    for age in mizrahi_data['ages']:
        # Calculate total population percentage for size reference
        miz_total = sum(mizrahi_data['populations'][age].values())
        ash_total = sum(ashkenazi_data['populations'][age].values())

        # Add Mizrahi data point
        data.append({
            'Age': age,
            'Group': 'Mizrahi',
            'Genetic Fit': mizrahi_data['genetic_fits'][mizrahi_data['ages'].index(age)],
            'Total Population': miz_total,
            'Dominant Population': max(mizrahi_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

        # Add Ashkenazi data point
        data.append({
            'Age': age,
            'Group': 'Ashkenazi',
            'Genetic Fit': ashkenazi_data['genetic_fits'][ashkenazi_data['ages'].index(age)],
            'Total Population': ash_total,
            'Dominant Population': max(ashkenazi_data['populations'][age].items(), key=lambda x: x[1])[0]
        })

    df = pd.DataFrame(data)

    fig = px.scatter(df,
                    x='Age',
                    y='Genetic Fit',
                    size='Total Population',
                    color='Group',
                    hover_data=['Dominant Population'],
                    title='Genetic Fits and Population Composition Over Time')

    return fig

def plot_all_visualizations():
    """Create and display all visualizations."""
    # Create bubble plot
    bubble_fig = create_bubble_plot(mizrahi_data, ashkenazi_data)
    bubble_fig.show()

    # Create visualizations for each time period
    for age in mizrahi_data['ages']:
        # Radar plot
        radar_fig = create_radar_plot(mizrahi_data, ashkenazi_data, age)
        radar_fig.show()

        # Sankey diagram
        sankey_fig = create_sankey_diagram(mizrahi_data, ashkenazi_data, age)
        sankey_fig.show()

if __name__ == "__main__":
    # Make sure to install required packages
    # !pip install plotly
    plot_all_visualizations()
```

## Bronze Age (approximately 3000-1200 BCE)

### Mizrahi Jews
- **Primary components**: Bactria-Margiana (31.0%), Eastern Steppe (25.2%), Central Steppe (23.6%), Yellow River (15.0%)
- The Mizrahi profile shows strong Central Asian influence with significant Eastern connections.
- The genetic fit score (2.572) suggests relatively good alignment with these source populations.

### Ashkenazi and Sephardic Jews (identical Bronze Age profiles)
- **Primary components**: Eastern Steppe (29.0%), Western Steppe (26.8%), Bronze Age Caucasian (21.4%), Yellow River (15.2%)
- Both groups show a steppe-dominant ancestry with Caucasian influence, suggesting a more western position than Mizrahi Jews during this period.
- Small Canaanite component (2.4%) indicates early Levantine connections.
- The genetic fit score (3.789) is higher than for Mizrahi, indicating less precise matching to these ancestral populations.

**Bronze Age Summary**: During this period, the ancestors of these Jewish populations were not yet culturally or genetically unified. The data suggests that Mizrahi Jews' ancestors were positioned further east in Central Asia, while Ashkenazi and Sephardic ancestors shared common roots in the steppe regions with Caucasian influence. All three groups show Yellow River (East Asian) components (15.0-15.2%), indicating extensive trade or migration networks.

## Iron Age (approximately 1200-300 BCE)

### Mizrahi Jews
- **Primary components**: Xiongnu (43.4%), Yaz Culture (20.0%), Mannaean (18.8%), Tarim Basin (16.6%)
- Strong Central Asian nomadic influence with growing Iranian connections.
- The emergence of the Tarim Basin component suggests eastward connections along early Silk Road routes.

### Ashkenazi Jews
- **Primary components**: Xiongnu (47.2%), Colchian (31.2%), Balto-Slavic (13.8%), Sinitic (4.8%)
- Retains strong Central Asian nomadic influence but shows significant Black Sea/Caucasus connections (Colchian).
- Early Balto-Slavic component indicates northward movements into Eastern Europe.

### Sephardic Jews
- **Primary components**: Xiongnu (53.4%), Colchian (34.6%), Italic and Etruscan (9.0%)
- Highest Xiongnu component of all three groups.
- The Italic and Etruscan component (instead of Balto-Slavic) suggests different westward migration patterns than Ashkenazi Jews.

**Iron Age Summary**: All three groups show strong Xiongnu influence, suggesting that Central Asian nomadic expansions affected ancestral Jewish populations broadly. However, the secondary components reveal diverging paths: Mizrahi ancestors maintained connections to Iranian plateau and Central Asian civilizations, while Ashkenazi and Sephardic ancestors began moving westward through the Caucasus region, with early differentiation in their respective westward trajectories.

## Late Antiquity (approximately 300 BCE-500 CE)

### Mizrahi Jews
- **Primary components**: Khotanese Saka (46.6%), Rouran Khaganate (22.6%), Hunnic (15.4%), Iranian Plateau (8.6%)
- Firmly established in Central Asia with strong Iranian connections and eastern trade route influences.
- First appearance of Roman Levant component (5.8%), indicating some connection to the historical Jewish homeland.

### Ashkenazi and Sephardic Jews (identical Late Antiquity profiles)
- **Primary components**: Hunnic (54.2%), Roman Levant (26.2%), Slavic (15.2%), Sinitic (4.4%)
- Much stronger Roman Levant component than Mizrahi Jews, indicating closer ties to the historical Jewish centers.
- The high Hunnic component suggests influence from these Central Asian nomads as they moved westward into Europe.
- The Slavic component points to continued settlement in Eastern Europe.

**Late Antiquity Summary**: This period shows the beginning of recognizable Jewish community formation. The Roman Levant component (26.2% in Ashkenazi/Sephardic vs. 5.8% in Mizrahi) suggests variable connections to the historical Jewish homeland in the Levant. The Mizrahi population remained more strongly connected to Central Asian and Iranian populations, while the Ashkenazi and Sephardic ancestors shared a common profile influenced by western migration of Hunnic peoples into Eastern Europe.

## Middle Ages (approximately 500-1500 CE)

### Mizrahi Jews
- **Primary components**: Turkic (89.6%), Indian Subcontinent (9.4%)
- Dramatic shift to Turkic dominance reflects Islamic conquests and Turkification of Central Asia.
- Indian Subcontinent component suggests trade connections along southern routes.
- Best genetic fit score (2.461) of all periods indicates strong alignment with these medieval populations.

### Ashkenazi Jews
- **Primary components**: Slavic (44.8%), European Jew (44.6%), Germanic (10.6%)
- The emergence of the European Jew component marks the crystallization of distinct Ashkenazi identity.
- Strong Slavic component reflects extensive settlement in Eastern Europe.
- Very high genetic fit score (27.080) suggests substantial genetic drift or founder effects in this population.

### Sephardic Jews
- **Primary components**: Slavic (66.2%), North African (26.4%), European Jew (7.4%)
- Higher Slavic component than Ashkenazi Jews, possibly indicating initial Eastern European settlement before westward movement.
- Significant North African component reflects migration to Iberia and North Africa following Islamic conquests.
- Very high genetic fit score (26.662) suggests strong founder effects similar to Ashkenazi Jews.

**Middle Ages Summary**: This period marks the clearest differentiation between the three Jewish populations. Mizrahi Jews became thoroughly integrated into the Turkic-dominated societies of the Middle East and Central Asia. Ashkenazi Jews established their distinct identity in Eastern and Central Europe with approximately equal Slavic and European Jewish components. Sephardic Jews show a trajectory through Eastern Europe but with significant North African influence, reflecting their historical presence in Iberia and North Africa during the Islamic Golden Age.

## Overall Migration Narrative

1. **Bronze Age**: All three populations show different positions along an east-west gradient, with Mizrahi ancestors further east in Central Asia and Ashkenazi/Sephardic ancestors sharing a common western steppe heritage.

2. **Iron Age**: While all groups show strong Central Asian nomadic (Xiongnu) influence, their secondary components reveal diverging paths: Mizrahi maintaining Central Asian and Iranian connections, Ashkenazi developing Balto-Slavic connections, and Sephardic showing Mediterranean (Italic/Etruscan) influence.

3. **Late Antiquity**: This period shows increasing connections to the Roman Levant (historical Jewish homeland) for Ashkenazi and Sephardic populations, while Mizrahi Jews maintained stronger Central Asian connections with minimal Levantine influence.

4. **Middle Ages**: The three populations crystallized into their recognizable historical forms - Mizrahi Jews thoroughly integrated into Turkic-dominated Middle Eastern societies, Ashkenazi Jews established in Eastern/Central Europe with a distinct genetic identity, and Sephardic Jews showing North African influence reflecting their Iberian and North African diaspora.

The dramatically higher genetic fit scores for Ashkenazi and Sephardic Jews in the Middle Ages (27.080 and 26.662 respectively) compared to earlier periods suggests significant founder effects and/or genetic drift, likely due to historical bottlenecks, endogamy, and isolation experienced by these communities in medieval Europe.
